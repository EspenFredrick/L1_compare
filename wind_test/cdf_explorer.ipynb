```python
ace_mfi = pd.read_csv('AC_H0_MFI_3649384.csv', header=0, parse_dates=[0], names=['Date', 'Bx', 'By', 'Bz', 'X', 'Y', 'Z'])
```

```python
B = ace_mfi[['Bx', 'By', 'Bz']].to_numpy()
```

```python
cdf = pycdf.CDF('ac_mfi_fromcsv.cdf', '')
cdf['Epoch'] = ace_mfi['Date']
cdf['B'] = B
cdf.attrs['Author'] = 'Espen Fredrick'
cdf.attrs['CreateDate'] = datetime.now()
cdf['B'].attrs['units'] = 'nT'
cdf.close()
```
